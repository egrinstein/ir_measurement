```python


import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
from scipy.fft import rfft, rfftfreq

SAMPLE_RATE = 44100

path_0 = "recorded/newRir7/RIR.npy"
path_1 = "recorded/newRir8/RIR.npy"

path_2 = "recorded/newRir9/RIR.npy"

path_3 = "recorded/newRir13/RIR.npy"

def plot_rir(path):
    rir = np.load(path)
    print(rir.shape)
    fig, axs = plt.subplots(nrows=2)
    axs[0].plot(rir)
    axs[0].set_title("Impulse responses")

    transfer_func = rfft(rir, axis=0)
    xf = rfftfreq(rir.shape[0], 1 / SAMPLE_RATE)
    axs[1].plot(xf, np.abs(transfer_func)[:, 0])
    axs[1].set_title("Transfer functions")
    plt.tight_layout()

# plot_rir(path_0)
# plot_rir(path_1)
# plot_rir(path_2)
plot_rir(path_1)
```
